```python
import matplotlib.pyplot as pp
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('new.csv')
```

```python
x = data.x.to_numpy()
y = data.y.to_numpy()
```

```python
pp.scatter(x,y)
```

```python
x=x/500.0
y=y/500.0 #dividing each number by 500
```

```python
m,c = np.random.random(),np.random.random()
```

```python
def learn(x,y,m,c,epoch):
 for i in range(epoch):
    error = y - (m*x+c)
    x_error = (y - (m*x+c))*x #number wise
    error = np.sum(error)/len(x)
    x_error = np.sum(x_error)/len(x)
    learning_rate = 0.01
    delta_m=learning_rate *x_error
    delta_c=learning_rate*error
    m +=m+delta_m
    c +=c+delta_c
 return (m,c)

 


```

```python
m,c = learn(x,y,m,c,20000)
```

```python
x1 = np.array([0, 1])
y1 = m * x1 + c
```

```python
pp.scatter(x,y)
pp.plot(x1, y1)

```
